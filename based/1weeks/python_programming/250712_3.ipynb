# 데이터 시각화
- 시각화 라이브러리 호출
- x축, y축에 표시할 데이터 정하기
- plot()에 데이터 입력
- 그래프 출력

## 시각화 라이브러리
| 라이브러리   | 특징                                                                 |
|--------------|----------------------------------------------------------------------|
| Matplotlib   | 데이터를 그래프나 플롯으로 시각화하는 라이브러리                     |
| Seaborn      | Matplotlib 기반으로 색 테마, 그래프 기능 등을 추가해주는 라이브러리    |
| Plotly       | 상호작용용 그래프를 그려주는 라이브러리                              |
| Bokeh        | Matplotlib의 대화형 버전으로 다양한 기능을 제공                      |
| Folium       | 지도 데이터를 이용하여 위치 정보를 시각화하는 라이브러리              |



```python
# 기본 시각화 문법 = matplotlib 라이브러리 그냥 사용
import matplotlib.pyplot as plt

x = [3, 5, 7, 9]
y = [2, 4, 6, 8]
plt.plot(x, y)
plt.show()
```

## Matplotlib 라이브러리
- 데이터를 그래프나 플룻으로 시각화 하는 라이브러리

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.title('Basic Line')
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# 막대 그래프
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

plt.bar(categories, values, color='red')
plt.title('Basic Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# 파이차트
import matplotlib.pyplot as plt

labels = ['apple', 'banana', 'cherry', 'pear']
size = [30, 25, 25, 20]

plt.pie(size, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie chart')
plt.show()
```

```python
# 여러 그래프 출력 - subplot

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r--')
plt.title('Line')

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='cyan')
plt.title('Bar')

plt.show()
```

```python
# 그래프 스타일 - 선, 마커, 색상 등 - plot 차트 활용
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, color='purple', linestyle='-.', marker='o', markersize=8)
plt.title('Line Style')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Seaborn 라이브러리
- Matplotlib 기반으로 색 테마, 그래프 기능 등을 추가해주는 라이브러리
- Matplotlib에 다양한 테마 및 통계용 그래프 등 동적인 기능을 추가한 시각화 라이브러리
- Matplotlib에 없는 히트맵, 카운트플롯 등다양한 종류 그래프 지원

```python
# 라이브러리 임포트 및 데이터 로드

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# 로컬인경우 표현하실분
path = 'week2_data'
# 코랩인경우
# path = '/content/drive/MyDrive/★ 클뉴즈sw/인공지능(AI) 교육/코랩실습/2주차/week2_data'

# 엑셀 데이터 불러오기
df = pd.read_excel(f'{path}/seaborn_data.xlsx')
# df


# 성별의 따른 구매 구매금액 분포
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='PurchaseAmount', data=df)
plt.title('Purchase Amount by Gender')
plt.show()
```

```python
# 라이브러리 임포트 및 데이터 로드

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# 로컬인경우 표현하실분
path = 'week2_data'
# 코랩인경우
# path = '/content/drive/MyDrive/★ 클뉴즈sw/인공지능(AI) 교육/코랩실습/2주차/week2_data'

# 엑셀 데이터 불러오기
df = pd.read_excel(f'{path}/seaborn_data.xlsx')
# df


# 구매상품별 평균 구매금액
plt.figure(figsize=(8, 6))
# declear때문에 - ci부분, errorbar=None으로 대처하라고 유도해서
# sns.barplot(x='Product', y='PurchaseAmount', data=df, estimator=sum, ci=None)
sns.barplot(x='Product', y='PurchaseAmount', data=df, estimator=sum, errorbar=None)
plt.title('Purchase Amount Distribution')
plt.xticks(rotation=45)
plt.show()
```
