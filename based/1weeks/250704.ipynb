## ✅ 벡터 투영 실습

- 벡터 `x = [1, 2]` 를 기준 벡터 `v = [1, 1]` 에 투영해보는 실습입니다.
- 이 과정에서 다음 개념을 다룹니다:
  - **벡터의 노름 (Norm)**: `l2 norm`, `l1 norm`
  - **내적 (Dot product)** 계산
  - **직교성(orthogonality)** 여부 확인
  - **벡터 투영 (Projection)** 구현 및 시각화




```python
import numpy as np
import matplotlib.pyplot as plt

# 벡터 정의
x = np.array([1, 2])
v = np.array([1, 1])  # 투영 기준 벡터

# 1. 노름
euclidean_norm = np.linalg.norm(x)
manhattan_norm = np.sum(np.abs(x))

# 2. 내적
dot_product = np.dot(x, v)

# 3. 직교성 확인
is_orthogonal = np.dot(x, v) == 0

# 4. 벡터 투영 (projection of x onto v)
def projection(x, v):
    return (np.dot(x, v) / np.dot(v, v)) * v

proj_x = projection(x, v)

# 5. 시각화
plt.quiver(0, 0, *x, angles='xy', scale_units='xy', scale=1, color='g', label='x')
plt.quiver(0, 0, *v, angles='xy', scale_units='xy', scale=1, color='b', label='v')
plt.quiver(0, 0, *proj_x, angles='xy', scale_units='xy', scale=1, color='r', label='proj_v(x)')
plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.grid()
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.legend()
plt.title("Projection of x onto v")
plt.show()
```

**📐 Euclidean Norm (ℓ₂):**
$begin:math:display$
\\|x\\|_2 = \\sqrt{x_1^2 + x_2^2 + \\cdots + x_n^2} = \\sqrt{x^\\top x}
$end:math:display$

**📐 Manhattan Norm (ℓ₁):**
$begin:math:display$
\\|x\\|_1 = |x_1| + |x_2| + \\cdots + |x_n|
$end:math:display$
