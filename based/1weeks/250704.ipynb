## ‚úÖ Î≤°ÌÑ∞ Ìà¨ÏòÅ Ïã§Ïäµ

- Î≤°ÌÑ∞ `x = [1, 2]` Î•º Í∏∞Ï§Ä Î≤°ÌÑ∞ `v = [1, 1]` Ïóê Ìà¨ÏòÅÌï¥Î≥¥Îäî Ïã§ÏäµÏûÖÎãàÎã§.
- Ïù¥ Í≥ºÏ†ïÏóêÏÑú Îã§Ïùå Í∞úÎÖêÏùÑ Îã§Î£πÎãàÎã§:
  - **Î≤°ÌÑ∞Ïùò ÎÖ∏Î¶Ñ (Norm)**: `l2 norm`, `l1 norm`
  - **ÎÇ¥Ï†Å (Dot product)** Í≥ÑÏÇ∞
  - **ÏßÅÍµêÏÑ±(orthogonality)** Ïó¨Î∂Ä ÌôïÏù∏
  - **Î≤°ÌÑ∞ Ìà¨ÏòÅ (Projection)** Íµ¨ÌòÑ Î∞è ÏãúÍ∞ÅÌôî




```python
import numpy as np
import matplotlib.pyplot as plt

# Î≤°ÌÑ∞ Ï†ïÏùò
x = np.array([1, 2])
v = np.array([1, 1])  # Ìà¨ÏòÅ Í∏∞Ï§Ä Î≤°ÌÑ∞

# 1. ÎÖ∏Î¶Ñ
euclidean_norm = np.linalg.norm(x)
manhattan_norm = np.sum(np.abs(x))

# 2. ÎÇ¥Ï†Å
dot_product = np.dot(x, v)

# 3. ÏßÅÍµêÏÑ± ÌôïÏù∏
is_orthogonal = np.dot(x, v) == 0

# 4. Î≤°ÌÑ∞ Ìà¨ÏòÅ (projection of x onto v)
def projection(x, v):
    return (np.dot(x, v) / np.dot(v, v)) * v

proj_x = projection(x, v)

# 5. ÏãúÍ∞ÅÌôî
plt.quiver(0, 0, *x, angles='xy', scale_units='xy', scale=1, color='g', label='x')
plt.quiver(0, 0, *v, angles='xy', scale_units='xy', scale=1, color='b', label='v')
plt.quiver(0, 0, *proj_x, angles='xy', scale_units='xy', scale=1, color='r', label='proj_v(x)')
plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.grid()
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.legend()
plt.title("Projection of x onto v")
plt.show()
```
