### ✅ 베르누이 분포 (Bernoulli Distribution)
- 이산형(discrete) 확률 분포
- 결과 두가지 - boolean 형 - 성공/실패, 1/0
- 확률: 변수 X는 1의 확률이 p, 0의 확률이 1-p
- 수식:
  $$
  P(X = x) = p^x (1-p)^{1-x}, \quad x \in \{0, 1\}
  $$

```python
import numpy as np
import matplotlib.pyplot as plt

# 한글 폰트 설정  - MAC 기준
plt.rcParams['font.family'] = 'AppleGothic'  # 맥사용자
plt.rcParams['axes.unicode_minus'] =  False

# 베르누이 분포 (Bernoulli Destribution) 예시
# np.random.binomial = 이항 분포로 부터 무작위 표본 추출
p = 0.7
samples = np.random.binomial(1, p, size=1000)

# 시각화
plt.hist(samples, bins=2, rwidth=0.8)
plt.title("베르누이 분포 (p=0.7)")
plt.xlabel("값")
plt.ylabel("빈도")
plt.xticks([0, 1])
plt.show()
```

### ✅ 범주형 분포 (Categorical Distribution)
- 이산형 다항 분포 (카테고리 값 중 하나 선택)
- 예: 색상 (빨강, 초록, 파랑), 클래스 (A, B, C)
- 확률 변수 X는 k개의 카테고리 중 하나를 가짐
- 수식:
  $$
  P(X = i) = p_i, \quad \sum_{i=1}^{k} p_i = 1
  $$

```python
import numpy as np
import matplotlib.pyplot as plt

# 한글 폰트 설정  - MAC 기준
plt.rcParams['font.family'] = 'AppleGothic'  # 맥사용자
plt.rcParams['axes.unicode_minus'] =  False

# 범주형 분포 (Categorical Distribution)
categories = ['빨강', '초록', '파랑']
probs = [0.2, 0.5, 0.3]
sample = np.random.choice(categories, size=1000, p=probs)

# 시각화
plt.hist(samples, bins=np.arange(len(categories)+1)-0.5, rwidth=0.8)
plt.xticks(range(len(categories)), categories)
plt.title("범주형 분포 예시")
plt.xlabel("카테고리")
plt.ylabel("빈도")
plt.show()
```
