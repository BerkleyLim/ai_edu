# üìò 02. Ï°∞Í±¥Î∂Ä ÌôïÎ•† (Conditional Probability)

## ‚úÖ Ï†ïÏùò
Ïñ¥Îñ§ ÏÇ¨Í±¥ BÍ∞Ä ÏùºÏñ¥ÎÇ¨Îã§Îäî Ï°∞Í±¥ ÌïòÏóê ÏÇ¨Í±¥ AÍ∞Ä ÏùºÏñ¥ÎÇ† ÌôïÎ•†ÏùÑ Ï°∞Í±¥Î∂Ä ÌôïÎ•†Ïù¥ÎùºÍ≥† Ìï©ÎãàÎã§.

## ‚úÖ ÏàòÏãù
$$
P(A | B) = P(A ‚à© B) / P(B)
$$

Ï¶â, ÏÇ¨Í±¥ BÍ∞Ä Ï£ºÏñ¥ÏßÑ ÏÉÅÌô©ÏóêÏÑú AÏôÄ BÍ∞Ä ÎèôÏãúÏóê ÏùºÏñ¥ÎÇ† ÌôïÎ•†ÏùÑ BÍ∞Ä ÏùºÏñ¥ÎÇ† ÌôïÎ•†Î°ú ÎÇòÎàà Í∞íÏûÖÎãàÎã§.

```python
import numpy as np
import pandas as pd

# ÏãúÎìú Í≥†Ï†ï
np.random.seed(42)

# Îëê Î≥ÄÏàò X, Y ÏÉùÏÑ±
# X: 0.6 ÌôïÎ•†Î°ú 1, 0.4 ÌôïÎ•†Î°ú 0
# Y: XÍ∞Ä 1ÏùºÎïå, YÎäî 0.8 ÌôïÎ•†Î°ú 1
#    XÍ∞Ä 0ÏùºÎïå, YÎäî 0.3 ÌôïÎ•†Î°ú 1
X = np.random.choice(a=[0, 1], size=10000, p=[0.4, 0.6])
Y = np.array([np.random.choice([0, 1], p=[0.2, 0.8]) if x == 1 else np.random.choice([0, 1], p=[0.7, 0.3]) for x in X])

# ÌôïÎ•† Í≥ÑÏÇ∞
p_x1 = np.mean(X == 1)
p_y1 = np.mean(Y == 1)
p_y1_given_x1 = np.mean(Y[X == 1])   # Ï°∞Í±¥Î∂Ä ÌôïÎ•† P(Y=1 | X=1)
p_xy11 = np.mean((X == 1) & (Y == 1))  # P(X=1 and Y=1)
p_x1_y1 = p_xy11 / p_x1 # ÏàòÏãùÏúºÎ°ú Îã§Ïãú ÌôïÏù∏

# Í≤∞Í≥º Ï†ïÎ¶¨
df = pd.DataFrame({
    'Ìï≠Î™©': ['P(X=1)', 'P(Y=1)', 'P(Y=1 | X=1)', 'P(X=1 and Y=1)', 'P(Y=1 | X=1) = P(X=1 and Y=1) / P(X=1)'],
    'Í∞í': [p_x1, p_y1, p_y1_given_x1, p_xy11, p_x1_y1]
})

df
```
